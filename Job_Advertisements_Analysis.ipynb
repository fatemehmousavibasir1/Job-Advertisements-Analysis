## Job Advertisement Analysis

In this section, we will analyze the job advertisement data. This analysis will only cover a small portion of the insights hidden within the data.


```python
import pandas as pd
from datetime import datetime
```

```python

df = pd.read_csv("job_posts.csv")
df
```

### Preprocessing

```python
# remove rows wihout "Salary", "Level" and "Collaboration Type" 
df = df.dropna(subset=["Salary","Level","Collaboration Type"],how='all')
df
```

```python

df["Salary"]=df['Salary'].replace({'MT':''}, regex=True)
df["Salary"]=df['Salary'].replace({'<':''}, regex=True)
df["Salary"]=df['Salary'].replace({'>':''}, regex=True)
df
```

Fill the missing values in the Salary column by the average salary per level. Specifically, if the Salary for a Junior (J) advertisement is missing, fill it with the average salary for all juniors; if it's for a Senior (S) advertisement, fill it with the average salary for all seniors; similarly, for Middle-level (M) and Intern (I) advertisements, use the respective averages.

```python
import numpy as np
# fill nan values in "Salary"
df_salary=df.copy()
df_salary['Salary']=df_salary['Salary'].astype('Int32')
df_salary=df_salary.drop(columns=['Offers Remote','Publish Time','Title','Close Time','State',"Collaboration Type"])

grouped=df_salary.groupby('Level')['Salary'].apply(lambda x: x.mean())

junior_mean=grouped[1]
mid_mean=grouped[2]
senior_mean=grouped[3]
intern_mean=grouped[0]



df.loc[(df['Level'] == 'J') & df['Salary'].isna(), 'Salary'] = junior_mean
df.loc[(df['Level'] == 'M') & df['Salary'].isna(), 'Salary'] = mid_mean
df.loc[(df['Level'] == 'S') & df['Salary'].isna(), 'Salary'] = senior_mean
df.loc[(df['Level'] == 'I') & df['Salary'].isna(), 'Salary'] = intern_mean
df
```

# Step 2: Job Posting Time Distribution Analysis 
In this step, you need to categorize the 24-hour day into four time periods (dawn, morning, noon, night) and count how many job advertisements were posted in each period. 

```python
from datetime import time

df_copy3 = df.copy()
df_copy3['Publish Time'] = pd.to_datetime(df_copy3['Publish Time'])

dawn = time(hour=5, minute=59, second=59)
morning = time(hour=11, minute=59, second=59)
noon = time(hour=17, minute=59, second=59)

def discretize_time1(t):
    if t.time() <= dawn:
        return "dawn"
    elif t.time() <= morning:
        return "morning"
    elif t.time() <= noon:
        return "noon"
    else:
        return "night"

df_copy3['Time'] = df_copy3['Publish Time'].apply(discretize_time1)

discrete_time = df_copy3['Time'].value_counts()

discrete_time = discrete_time.sort_values(ascending=False)

discrete_time
```
